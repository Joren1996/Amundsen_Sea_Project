```python
fig=plt.figure(figsize=(20,15))
plt.rcParams.update({'font.size': 18})

maxlat=np.where(grid.lat_1d>-73.5)[0][0]
minlon=np.where(grid.lon_1d>-115)[0][0]
maxlon=np.where(grid.lon_1d>-98.5)[0][0]


#Show old shelf ice
temp=shelfice_old.copy()
temp[grid.land_mask]=np.nan
ax=plt.subplot(3,3,1)
plt.title('Shelfice \n Naughten et al., (2022)')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon], grid.lat_2d[:maxlat,minlon:maxlon], temp[:maxlat,minlon:maxlon], cmap='Blues_r', vmin=-1000, vmax=0)
ax.set_facecolor('grey')
plt.colorbar(label='Ice Draft [m]')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
#latlon_axes(ax=ax, x=grid.lon_2d[:maxlat,minlon:maxlon], y=grid.lat_2d[:maxlat,minlon:maxlon])

#Show New shelfice
temp=shelfice.copy()
temp[grid.land_mask]=np.nan
grounded=((shelfice_old!=0) & (shelfice==0))
grounded=grounded[:maxlat,minlon:maxlon]
ax=plt.subplot(3,3,2)
plt.title('Shelfice \n Advanced GL')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], np.where(grounded, np.ones(np.shape(grounded)), np.nan*np.ones(np.shape(grounded))), cmap='spring_r')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], np.where(grounded, np.nan*np.ones(np.shape(grounded)), temp[:maxlat,minlon:maxlon]), cmap='Blues_r', vmin=-1000, vmax=0)
plt.colorbar(label='Ice Draft [m]')
ax.set_facecolor('grey')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
#latlon_axes(ax=ax, x=grid.lon_2d[:maxlat,minlon:maxlon], y=grid.lat_2d[:maxlat,minlon:maxlon])

import matplotlib.patches as mpatches
yellow_patch = mpatches.Patch(facecolor='yellow', edgecolor='black', label='Newly Grounded Ice')

#Compute the differences
temp1=shelfice_old.copy()
temp2=shelfice.copy()
temp=temp2-temp1
temp[grid.land_mask]=np.nan
temp=temp[:maxlat,minlon:maxlon]

def newround(x, base=50):
    return base * round(x/base)

temp=np.where(grounded, np.nan*np.ones(np.shape(grounded)), temp)
divnorm = MidpointNormalize(vmin=np.nanmin(temp), vmax=np.nanmax(temp), midpoint=0)
ax=plt.subplot(3,3,3)
plt.title('Difference Shelfice \n Advanced GL - Naughten et al., (2022)')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], np.where(grounded, np.ones(np.shape(grounded)), np.nan*np.ones(np.shape(grounded))), cmap='spring_r')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], temp, cmap='seismic', norm=divnorm)
ax.set_facecolor('grey')
plt.colorbar(label='Ice Draft Difference [m]', ticks=np.arange(newround(np.nanmin(temp)), newround(np.nanmax(temp)), 100))
plt.xlabel('Longitude')
plt.ylabel('Latitude')
ax.set_facecolor('grey')
#latlon_axes(ax=ax, x=grid.lon_2d[:maxlat,minlon:maxlon], y=grid.lat_2d[:maxlat,minlon:maxlon])



#SHOW WATER COLUMN
#Show Old Water Column
temp=shelfice_old-bathy_old
temp[grid.land_mask]=np.nan
ax=plt.subplot(3,3,4)
plt.title('Water Column \n Naughten et al., (2022)')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon], grid.lat_2d[:maxlat,minlon:maxlon], temp[:maxlat,minlon:maxlon], cmap='rainbow', vmax=800)
ax.set_facecolor('grey')
plt.colorbar(label='Water Column [m]', extend='max')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
#latlon_axes(ax=ax, x=grid.lon_2d[:maxlat,minlon:maxlon], y=grid.lat_2d[:maxlat,minlon:maxlon])

#Show New Water Column
temp=shelfice-bathy
temp[grid.land_mask]=np.nan
grounded=((shelfice_old!=0) & (shelfice==0))
grounded=grounded[:maxlat,minlon:maxlon]
ax=plt.subplot(3,3,5)
plt.title('Water Column \n Advanced GL')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], \
           np.where(grounded, np.ones(np.shape(grounded)), np.nan*np.ones(np.shape(grounded))), cmap='spring_r')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], \
           np.where(grounded, np.nan*np.ones(np.shape(grounded)), temp[:maxlat,minlon:maxlon]), cmap='rainbow', vmax=800)
plt.colorbar(label='Water Column [m]', extend='max')
ax.set_facecolor('grey')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
#latlon_axes(ax=ax, x=grid.lon_2d[:maxlat,minlon:maxlon], y=grid.lat_2d[:maxlat,minlon:maxlon])

import matplotlib.patches as mpatches
yellow_patch = mpatches.Patch(facecolor='yellow', edgecolor='black', label='Newly Grounded')

#Compute the differences
temp1=shelfice_old-bathy_old
temp2=shelfice-bathy
temp=temp2-temp1
temp[grid.land_mask]=np.nan
temp=temp[:maxlat,minlon:maxlon]

def newround(x, base=50):
    return base * round(x/base)

temp=np.where(grounded, np.nan*np.ones(np.shape(grounded)), temp)
divnorm = MidpointNormalize(vmin=np.nanmin(temp), vmax=np.nanmax(temp), midpoint=0)
ax=plt.subplot(3,3,6)
plt.title('Difference Water Column \n Advanced GL - Naughten et al., (2022)')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], np.where(grounded, np.ones(np.shape(grounded)), np.nan*np.ones(np.shape(grounded))), cmap='spring_r')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], temp, cmap='seismic', norm=divnorm)
ax.set_facecolor('grey')
plt.colorbar(label='Water Column [m]', ticks=np.arange(newround(np.nanmin(temp)), newround(np.nanmax(temp)), 100))
plt.xlabel('Longitude')
plt.ylabel('Latitude')
ax.set_facecolor('grey')
#latlon_axes(ax=ax, x=grid.lon_2d[:maxlat,minlon:maxlon], y=grid.lat_2d[:maxlat,minlon:maxlon])


temp1=bathy_old.copy()
temp2=bathy.copy()
temp=temp2-temp1
temp[grid.land_mask]=np.nan
temp=temp[:maxlat,minlon:maxlon]

def newround(x, base=50):
    return base * round(x/base)

temp=np.where(grounded, np.nan*np.ones(np.shape(grounded)), temp)
divnorm = MidpointNormalize(vmin=np.nanmin(temp), vmax=np.nanmax(temp), midpoint=0)

ax=plt.subplot(3,2,5)
plt.title('Difference Bathymetry \n Advanced GL - Naughten et al., (2022)')
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], np.where(grounded, np.ones(np.shape(grounded)), np.nan*np.ones(np.shape(grounded))), cmap='spring_r')
#plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], np.where(grounded, np.nan*np.ones(np.shape(grounded)), temp), cmap='seismic', norm=divnorm)
plt.pcolor(grid.lon_2d[:maxlat,minlon:maxlon] , grid.lat_2d[:maxlat,minlon:maxlon], temp, cmap='seismic', norm=divnorm)
ax.set_facecolor('grey')
plt.colorbar(label='Bathymetry Difference [m]', ticks=np.arange(newround(np.nanmin(temp)), newround(np.nanmax(temp)), 100))
plt.xlabel('Longitude')
plt.ylabel('Latitude')
ax.set_facecolor('grey')
#latlon_axes(ax=ax, x=grid.lon_2d[:maxlat,minlon:maxlon], y=grid.lat_2d[:maxlat,minlon:maxlon])

ax=plt.subplot(3,2,6)
a=np.histogram(shelfice[:maxlat,minlon:maxlon].flatten(), bins=np.arange(-1000, 0, 20))
b=np.histogram(shelfice_old[:maxlat,minlon:maxlon].flatten(), bins=np.arange(-1000, 0, 20))
plt.plot(a[0]/np.sum(a[0]), a[1][:-1], 'r--o', label='Advanced GL')
plt.plot(b[0]/np.sum(b[0]), b[1][:-1], 'b--o', label='Naughten et \n al., 2022')
plt.ylabel('Depth [m]')
plt.xlabel('Fraction')
ax.set_yticks(np.arange(-1000, 1, 250), np.arange(0, 1001, 250)[::-1])
plt.legend(loc='lower right', fontsize=16)
plt.grid()
plt.title('Relative Distribution of \n Ice Draft Depth in Shown Domain')

#Add letters to subplot
import string
for n, ax in enumerate(fig.axes[::2]):
    ax.text(-0.25, 0.9, string.ascii_uppercase[n]+'.', transform=ax.transAxes, 
            size=25, weight='bold')

fig.legend(handles=[yellow_patch], loc='lower right')
plt.subplots_adjust(hspace=0.5, wspace=0.6)
fig.tight_layout()
fig.suptitle('Comparison between Geometry in Naughten et al., (2022) and Geometry with Advanced Grounding Line (GL)')

for ax in fig.axes[::2][:-1]:
    latlon_axes(ax=ax, x=grid.lon_2d[:maxlat,minlon:maxlon], y=grid.lat_2d[:maxlat,minlon:maxlon])

plt.savefig('/data/hpcdata/users/grejan/mitgcm/03_output/map_new_geometry.png')
```

#### Illustrate regions for climate indices

```python
plt.rcParams.update({'font.size': 20})
fig=plt.figure(figsize=(20,10))
ax = plt.subplot(111, projection=ccrs.PlateCarree(central_longitude=180.0))
ax.stock_img()

#Function to draw rectangles in the domain.
def plot_region(lat, lon, c, label=False):
    if type(label)==str:
        plt.plot([lon[-1], lon[-1], lon[0], lon[0], lon[-1]], \
                 [lat[0], lat[-1], lat[-1], lat[0], lat[0]], '-o', c=c, label=label,\
                transform=ccrs.PlateCarree(), lw=2)
    else:
        plt.plot([lon[-1], lon[-1], lon[0], lon[0], lon[-1]], \
                 [lat[0], lat[-1], lat[-1], lat[0], lat[0]], '-o', c=c,\
                transform=ccrs.PlateCarree(), lw=2)
#IPO
plot_region(lon=[140, 360-140], lat=[25, 45], c='b')
plot_region(lon=[170, 360-90], lat=[-10, 10], c='b')
plot_region(lon=[150, 360-160], lat=[-50, -15], c='b', label='IPO')

#SOI
plot_region(lon=[360-151, 360-149], lat=[-18.5, -17.5], c='g')
plot_region(lon=[130, 131], lat=[-13, -12], c='g', label='SOI')

#ASL
plot_region(lon=[170, 298], lat=[-80, -60], c='k', label='ASL')

#SAM
plot_region(lon=[2,359], lat=[-65,-65.3], c='r', label='SAM')
plot_region(lon=[2,359], lat=[-40,-40.5], c='r')

#WPSST
plot_region(lon=[120,160], lat=[0,15], c='y', label='WPSST')

ax.coastlines()
fig.suptitle('Areas used for Climate Indices')
ax.gridlines(draw_labels=True, crs=ccrs.PlateCarree())
    
ax.set_xlim([-179, 180])
ax.set_ylim([-90, 90])
ax.legend()

plt.savefig('/data/hpcdata/users/grejan/mitgcm/03_output/map_climate_indices.png')
```
